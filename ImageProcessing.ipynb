```python
#ImageProccessing
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import random
import imutils
```

```python
image= cv2.imread("Reiki_to_Mother_Earth.jpg")

cv2_imshow(image)
cv2.imwrite("Testing.jpg",image)
```

```python
can = np.zeros((1000,1000,3),dtype="uint8")*255
cv2_imshow(can)
```

```python
can = np.ones((1000,1000,3),dtype="uint8")*255
green = (0,255,0)
cv2.line(can,(0,0),(500,500),green, 3)

red = (0,100,255)
cv2.rectangle(can,(100,100),(300,300),red, -1)

cv2_imshow(can)
```

```python
#Create Circles
for i in range(0,25):
    radius=np.random.randint(5,100)
    color=np.random.randint(0,high=256,size=(3,)).tolist()
    point=np.random.randint(0,800,size=(2,))
    cv2.circle(can,list(point),radius,color,-1)
 
cv2.putText(can,"My Happy Circles",(100,100),cv2.FONT_HERSHEY_SIMPLEX,1.5,(100,0,50),5)
cv2_imshow(can)
```
